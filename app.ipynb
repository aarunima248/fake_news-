```python
def clean_text(text):
    text = text.lower()
    text = re.sub('\[.*?\]',"",text)
    text = re.sub("\\W"," ",text)
    text = re.sub("https?:://\S+|www|.\S+","",text)
    text = re.sub("<.*?>+","",text)
    text = re.sub("[%s] % re.escape(string.punctuation),"",text)
    text = re.sub("\n","",text)
    text = re.sub("\W*\d\W*","",text)
    return text
    
```
